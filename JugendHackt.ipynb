# plotting a basic sine wave

```python
import matplotlib.pyplot as plt # for ploting
import numpy as np 

fs = 100 # this is the sample rate 
f = 2 # the frequency of the signal

x = np.arange(fs) # the points on the x axis for plotting
# compute the value (the amplitude) of the sine wave at the for each sample
y = [ np.sin(2*np.pi*f * (i/fs)) for i in x]

% matplotlib inline
# showing the exact location of the samples
plt.stem(x,y, 'r', )
plt.plot(x,y)
```

# lets now plot a cosine wave, a wave with a phase-shift of Ï€/2 radians

```python
import matplotlib.pyplot as plt # for ploting
import numpy as np # to work with numerical data efficiently

fs = 100 # sample rate 
f = 2 # the frequency of the signal

x = np.arange(fs) # the points on the x axis for plotting
# compute the value (amplitude) of the cosine wave at the for each sample
y = [ np.cos(2*np.pi*f * (i/fs)) for i in x]

% matplotlib inline
# showing the exact location of the samples
plt.stem(x,y, 'r', )
plt.plot(x,y)
```

# signals can be evaluated at any point. feel free to change the frequency value.

people can hear sounds at frequencies ranging about 20 Hz to 20,000 Hz, though we hear sounds best from 1,000 Hz to 5,000 Hz, where human speech is centered. hearing loss may reduce the range of frequencies a person can hear. It is common for people to lose their ability to hear higher frequencies as they get older.

In music, an "A" translates to one of the following frequencies:

432   434 	 436 	 438 	 440 	 442 	 444 	 446 

440 is most comonly used. However when playing barock music, musicians tend to tune to as low as 430 Hz.


```python
cos_sig = thinkdsp.CosSignal(freq=440)
```

# a cosine signal at 440 Hz with a period of 2.3 ms

```python
cos_sig.plot()
thinkplot.config(xlabel='time (s)', legend=False)
```

# "make_wave" samples the signal at equally-space time steps.

```python
wave = cos_sig.make_wave(duration=0.5, framerate=11025)
```

# "make_audio" creates a widget that plays the wave. feel free to change the frequency in the above cell.

```python
wave.apodize()
wave.make_audio()
```

# "make_spectrum" returns a spectrum object. below we can see that a cosine wave contains only one frequency component (that means there are no harmonics).

```python
spectrum = wave.make_spectrum()

spectrum.plot()
thinkplot.config(xlabel='frequency (Hz)', legend=False)
```

# ofcourse we can also plot cool sawtooth signals. a so called sawtooths have a more complex harmonic structure. check it out:

the sawtooth wave is a really cool kind of non-sinusoidal waveform. it is so named based on its resemblance to the teeth of a plain-toothed saw with a zero rake angle.

a sawtooth wave ramps upward and then sharply drops. However, in a "reverse (or inverse) sawtooth wave", the wave ramps downward and then sharply rises. it can also be considered the extreme case of an asymmetric triangle wave.


```python
saw_sig = thinkdsp.SawtoothSignal(freq=200)
saw_sig.plot()
```

# here's what that weird function sounds like

```python
saw_wave = saw_sig.make_wave(duration=0.5)
saw_wave.make_audio()
```

# and here's what the spectrum looks like. feel free to play around with the frequency values. 


```python
saw_wave.make_spectrum().plot()
```
